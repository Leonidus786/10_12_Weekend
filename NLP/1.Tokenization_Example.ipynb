```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
```

```python
print(documents)
```

```python
for sentence in documents:
    print(sentence)
```
